```python
# -------------------------------------------------
# PGF backend (MUST be before pyplot import)
# -------------------------------------------------
import matplotlib
matplotlib.use("pgf")

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# -------------------------------------------------
# LaTeX configuration
# -------------------------------------------------
plt.rcParams.update({
    "pgf.texsystem": "pdflatex",     # or "lualatex", "xelatex"
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": [],                # use LaTeX default (Computer Modern)
    "pgf.rcfonts": False,
    "axes.unicode_minus": False,
})

# -------------------------------------------------
# Models & data
# -------------------------------------------------
models = [
    "DistilBERT (Ours)",
    "Bertha von Suttner",
    "Fine-Tuned DistilBERT (Ours)",
    "GPT-4.1-mini",
    "GloVe + CNN (Ours)",
]

metric_names = ["Accuracy", "Precision", "Recall", "F1"]

data = np.array([
    [0.828, 0.801, 0.873, 0.835],
    [0.822, 0.871, 0.755, 0.809],
    [0.807, 0.856, 0.739, 0.793],
    [0.807, 0.785, 0.847, 0.815],
    [0.725, 0.793, 0.608, 0.668],
])

# -------------------------------------------------
# Plot setup
# -------------------------------------------------
x = np.arange(len(metric_names))
width = 0.15

fig, ax = plt.subplots(figsize=(14, 8))

colors = ["#4C6CFE", "#BFBFBF", "#ABC3FF", "#E1E1E1", "#D6E4FF"]
hatches = ["//", "", "", "", ""]

# -------------------------------------------------
# Bars
# -------------------------------------------------
for i, model in enumerate(models):
    bars = ax.bar(
        x + (i - 2) * width,
        data[i],
        width,
        color=colors[i],
        hatch=hatches[i],
        edgecolor="white",
        linewidth=0.6,
    )

    # Value labels
    for bar in bars:
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.008,
            f"{bar.get_height():.3f}",
            ha="center",
            va="bottom",
            fontsize=9,
        )

# -------------------------------------------------
# Axes styling
# -------------------------------------------------
ax.set_ylabel("Score", fontsize=12)
ax.set_ylim(0.6, 0.9)
ax.set_xticks(x)
ax.set_xticklabels(metric_names, fontsize=12)

ax.grid(axis="y", linestyle="--", linewidth=0.6, alpha=0.35)
ax.set_axisbelow(True)

# Chart border
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.0)
    spine.set_edgecolor("#CFCFCF")

# -------------------------------------------------
# Legend
# -------------------------------------------------
legend_handles = [
    Patch(
        facecolor=colors[i],
        hatch=hatches[i],
        edgecolor="#CFCFCF",
        linewidth=0.8,
        label=models[i],
    )
    for i in range(len(models))
]

legend = ax.legend(
    handles=legend_handles,
    loc="upper center",
    bbox_to_anchor=(0.5, 1.10),
    ncol=5,
    fontsize=11,
    handlelength=1.8,
    columnspacing=1.6,
    frameon=True,
    fancybox=True,
)

legend.get_frame().set_edgecolor("#D9D9D9")
legend.get_frame().set_linewidth(1.0)
legend.get_frame().set_facecolor("white")
legend.get_frame().set_alpha(1.0)

# -------------------------------------------------
# Layout & export
# -------------------------------------------------
plt.tight_layout()
plt.savefig("model_comparison.pgf")
plt.close()
```
